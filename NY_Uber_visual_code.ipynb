```python
df.columns = ['day', 'temp_h', 'temp_a', 'temp_l', 'dew_h', 'dew_a', 'dew_l', 'hun_h', 'hum_a', 'hum_l', 'seap_h', 'seap_a', 'seap_l', 'vis_h', 'vis_a', 'vis_l', 'win_h', 'win_a', 'pre_h', 'sum', 'rain']
```

## Eliminate Redundant Rows

```python
df[df['hum_a'].str.contains('')==True]
```

```python
df = df.drop(df[df['hum_a'].str.contains('')==True].index)
```

```python
df
```

```python
df.index = range(122)
```

```python
df['rain'] = df['rain'].fillna('Clear')
```

## Assign Values for Month

```python
df["month"] = ""
```

```python
df['month'][:30] = 'April'
df['month'][30:61] = 'May'
df['month'][61:91] = 'June'
df['month'][91:122] = 'July'
```

```python
df
```

```python
april = df[:30]
```

```python
may = df[30:61]
```

```python
june = df[61:91]
```

## Join weather and uber tables

```python
ub1 = pd.read_csv('C:/Users/james/Desktop/Wecloud_Data_Bootcamp/uber_tableau_challenge/uber-raw-data-apr14.csv')
```

```python
ub1
```

```python
ub1['datetime'] = pd.to_datetime(ub1['Date/Time'])

ub1['date'] = ub1['datetime'].dt.date
ub1['month'] = ub1['datetime'].dt.month
ub1['day'] = ub1['datetime'].dt.day
ub1['hour'] = ub1['datetime'].dt.hour
```

```python
ub1
```

```python
df1 = april.merge(ub1,how='right',on='day').drop(columns=['Date/Time', 'sum', 'win_h', 'win_a'])

df1
```

```python
ub2 = pd.read_csv('C:/Users/james/Desktop/Wecloud_Data_Bootcamp/uber_tableau_challenge/uber-raw-data-may14.csv')
```

```python
ub2['datetime'] = pd.to_datetime(ub2['Date/Time'])

ub2['date'] = ub2['datetime'].dt.date
ub2['month'] = ub2['datetime'].dt.month
ub2['day'] = ub2['datetime'].dt.day
ub2['hour'] = ub2['datetime'].dt.hour
ub2
```

```python
df2 = may.merge(ub2,how='right',on='day').drop(columns=['Date/Time', 'sum', 'win_h', 'win_a'])

df2
```

```python
ub3 = pd.read_csv('C:/Users/james/Desktop/Wecloud_Data_Bootcamp/uber_tableau_challenge/uber-raw-data-jun14.csv')
```

```python
ub3['datetime'] = pd.to_datetime(ub3['Date/Time'])

ub3['date'] = ub3['datetime'].dt.date
ub3['month'] = ub3['datetime'].dt.month
ub3['day'] = ub3['datetime'].dt.day
ub3['hour'] = ub3['datetime'].dt.hour
ub3
```

```python
df3 = june.merge(ub3,how='right',on='day').drop(columns=['Date/Time', 'sum', 'win_h', 'win_a'])

df3
```

```python
frames = [df1, df2, df3]
```

```python
df_result = pd.concat(frames)
```
